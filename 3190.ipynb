```python
# check board
for i in range(1, N+1):
    print(board[i][1:])

direction = [[0, 1], [1, 0], [0, -1], [-1, 0]] # R => +1, L => -1

from collections import deque

class Snake:
    def __init__(self, length=1, bodies=[[1,1]], bodies_dir = [0], dir=0):
        self.length = length
        self.bodies = bodies # [y,x, dir]
        board[self.bodies[0][0]][self.bodies[0][1]] = +100
        self.bodies_dir = bodies_dir        
        self.dir = dir
    
    def move(self, nd):
        # params nd : next direction
        self.dir = (self.dir + nd)%4
        nd = self.dir
        tail = self.bodies[-1]
        last_dir = self.bodies_dir[-1]

        for i in range(self.length):
            [y, x] = self.bodies[i]
            board[y][x] -= 100
            ny = y+direction[nd][0]; nx = x+direction[nd][1]
            if ny <= 0 or nx <= 0 or ny > N or nx > N:
                return None, None

            if board[ny][nx] >= 100: # collapse
                return None, None
            self.bodies[i] = [ny, nx]
            board[ny][nx] += 100

            next_direction = self.bodies_dir[i]
            self.bodies_dir[i] = nd
            nd = next_direction
        return tail, last_dir # tail

    def add_tail(self, tail, last_dir):
        self.bodies.append(tail); board[tail[0]][tail[1]] = 100
        self.bodies_dir.append(last_dir)
        self.length += 1
```

```python
def move(snake, movements, stop=-1):
    global N, L
    time = 0
    next_move = 0

    while True:
        if time == stop:
            return time

        # all move or next direction
        if next_move >= L or int(movements[next_move][0]) != time:
            nd = 0
        else: # int(movements[next_move][0]) == time:
            nd = movements[next_move][1]
            next_move += 1

        if nd == 'D': # right turn
            nd = 1
        elif nd == 'L': # left turn
            nd = -1
        else: # no turn. nd = 0
            pass

        tail, last_dir = snake.move(nd) # move snake

        time += 1
        if tail is None or last_dir is None: # collapse or out 
            return time
        
        if board[snake.bodies[0][0]][snake.bodies[0][1]]%100 == 1: # is apple
            print("EAT APPLE")
            board[snake.bodies[0][0]][snake.bodies[0][1]] -= 1 # eat apple
            snake.add_tail(tail, last_dir)
        else: # no apple
            pass
    return time
```
